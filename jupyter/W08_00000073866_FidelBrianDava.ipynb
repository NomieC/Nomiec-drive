```python
import numpy as np
import matplotlib.pyplot as plt

#Define π & number of points
π = np.pi
num_points = 400

#X values from -π to π
x = np.linspace(-π, π, num_points)

#Calculate function values
y = np.sin(x)
y_approx_linear = x
y_div_x = np.sin(x) / x

#Print each (x, y) pair for approximation
for i in range(len(x)):
    print(f"For x = {x[i]:.16f}, y = {y[i]:.16f}, y_approx_linear = {y_approx_linear[i]:.16f}, y_div_x = {y_div_x[i]:.16f}")

#Plot functions and approximations
plt.plot(x, y, label="sin(x) or Analytic")
plt.plot(x, y_approx_linear, label="Linear Approximation")
plt.plot(x, y_div_x, label="sin(x) / x")
plt.plot(x, y_div_x, linestyle='--', label="sin(x) / x Approximation")

#Set labels & title
plt.xlabel("x")
plt.ylabel("y")
title = "Taylor Series Approximations of sin(x) (a=0)"
plt.title(title)
plt.legend()

#Plot with grid
plt.grid(True)
plt.show()
```
